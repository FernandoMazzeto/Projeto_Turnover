# **Diminuição de Turnover de Colaboradores de uma Empresa de Tecnologia**

## **1. Introdução**

### **1.1. Contexto**

O projeto está associado com a diminuição de Turnover dos colaboradores de uma determinada empresa de tecnologia. Nesse sentido, o Turnover (também apresentado como rotatividade dos funcionários) é um grande problema para as empresas. Sempre que um funcionário deixa um determinado trabalho, a empresa perde dinheiro e tempo com novas entrevistas e treinamentos do novo funcionário. Isso sem falar da perda de produtividade do setor afetado por essa rotatividade. Além disso, inúmeros fatores podem ocasionar a perda de um colaborador, assim, o presente projeto busca entender, contextualizar e solucionar os fatores associados a essa determinada empresa.

### **1.2. Objetivo**

O projeto apresenta como objetivo contextualizar e desenvolver as políticas ou fatores da empresa que devem ser alteradas de forma a minimizar o turnover e auxiliar na melhor organização da empresa.

### **1.3. Visão Regulatória**

A visão regulatória tem como objetivo assegurar que todo o processo de turnover dos colaboradores está em total conformidade com as legislações trabalhistas, proteção de dados e fiscais.





#### **1.3.1. Legislação Trabalhista**

##### **1.3.1.1. Consolidação das Leis do Trabalho (CLT)**

**Lei nº 5.452/1943:** Estabelece os direitos trabalhistas, incluindo regras para rescisão contratual, aviso prévio, pagamento de verbas rescisórias e estabilidade de emprego.

#### **1.3.2. Normas Previdenciárias e Fiscais**

##### **1.3.2.1. Portaria SEPRT nº 1.195/2019:**

Regula o recolhimento de FGTS e INSS nos casos de desligamentos, especificando prazos e valores.

#### **1.3.3 Proteção de Dados e Privacidade**

##### **1.3.3.1. Lei Geral de Proteção de Dados (LGPD)**

**Lei nº 13.709/2018:** Regula o tratamento de dados pessoais, exigindo consentimento para coleta e uso de dados de funcionários e garantindo o princípio da transparência nos processos de turnover.

#### **2.3.2. Relação Variável e Turnover**

Para determinar a força e a influência de cada variável em relação ao turnover dos colaboradores, foi utilizada a técnica conhecida como Information Value (IV). Essa técnica estatística é utilizada para quantificar e demonstrar a importância de uma variável na previsão de um resultado. Desta forma, ela auxilia em diversos aspectos da análise, tais como: na identificação de variáveis relevantes e a redução de dimensionalidade. #### **2.3.3. Variáveis de Destaque**

A partir da tabela de forças desenvolvida, com o intuito de objetividade e melhor desenvolvimento do projeto, a escolha é de focar nos aspectos apresentados pelas variáveis que possuem um forte poder de separabilidade, ou seja, as variáveis: **Faz_hora_extras?, Salário, Idade e Tempo_de_carreira.**

##### **2.3.3.1. Faz_hora_extras?**


*   O número de colaboradores que não fazem hora extra é de **1.054** e os que fazem hora extra correspondem a **416** colaboradores;
*   Dos 1.054 colaboradores que não fazem hora extra, realizaram o processo de turnover apenas **110**, o que correspoonde a **10%** dos que não fazem hora extra;
*   Dos 416 que fazem hora extra, realizaram o processo de turnover 127 colaboradores, o que corresponde a **31%** dos que fazem hora extra.

É possível analisar que a realização de horas extras impacta diretamente no turnover de funcionários, visto que, proporccionalmente, os indivíduos que realizam hora extra apresentam um número considerável de turnover.

##### **2.3.3.2. Salário**



* O número de colaboradores com salários entre 1.009 e 4.008 reiais é de **544**, enquanto aqueles com salários entre 16.009 e 19.008 reais somam apenas **76** colaboradores.
*   A faixa salarial de 16.009 a 19.008 reais apresenta a menor taxa de turnover, com valores próximos a **1%** do total de colaboradores nessa faixa.
*   Já a faixa salarial de 1.009 a 4.008 reais possui a maior taxa de turnover, atingindo aproximadamente **25%** do total de colaboradores nessa faixa.
*   A segunda pior taxa de turnover é observada na faixa salarial entre 10.009 e 13.008 reais, com um índice de **16%** do total de colaboradores nessa faixa.

Desta forma, pode-se perceber que a faixa que apresenta o maior problema de turnover é a que representa os menores salários da empresa.

##### **2.3.3.3. Idade**



*   O número de colaboradores com idade entre 18 e 22 anos é de **57**, já para a faixa entre 38 e 42 anos é de **243** colaboradores;
*   A faixa de idade entre 38 e 42 anos apresenta a menor taxa de turnover da empresa, com apenas **9%** dos colaboradores dessa faixa;
*   Já, para os colaboradores de idade entre 18 e 22 anos apresenta a pior taxa de turnover com larga distância para as outras faixas, representando cerca de **47%** dos colaboradores dessa faixa;

Dessa forma, é possível perceber um grande problema para a empresa com relação aos colaboradores mais jovens, visto que apresenta uma alta e impactante taxa de turnover.

##### **2.3.3.4. Tempo de Carreira**



*   O número de colaboradores com tempo de carreira entre 0 e 4 anos é de **228**, já para os colaboradores com tempo de carreira superior a 20 anos é de **207** colaboradores;
*  Os colaboradores com, no máximo, 4 anos de carreira representem a pior taxa de turnover, com cerca de **33%** do total dessa faixa;
*   Já, a faixa superior a 20 anos, representa a melhor taxa de turnover, com apenas **8%** do total dessa faixa.

Assim, pode-se perceber que há um problema impactante na empresa, principalmente com os os indivíduos em início de carreira, uma vez que a taxa apresentada para essa faixa apresenta grande relevância e procupação. Porém, é necessário ressaltar que quanto maior o tempo de carreira, memor tende a ser a taxa de turnover dos colaboradores dessa empresa.
